```python
pickle_in = open("file_list.pickle","rb")
fn = pickle.load(pickle_in)
```

```python
(rate,sig) = wav.read((fn[0]))
```

```python
l6=len(mfcc(sig,rate)) # length 6
```

```python
def make3():
    data=[]
    label=[]
    for _ in range(3):
        r=random.randint(0,len(fn)-1)
        inf=fn[r]
        (rate,sig)=wav.read(inf)
        mfcct=mfcc(sig,rate)
        data=data+list(mfcct)
        cat=int((inf.split("/")[-2]).split("_")[0])
        le=[]
        for i in range(round(6*len(mfcct)/l6)):
            le=le+[cat]
        label=label+le
    return np.array(data),label
```

```python
def make2():
    data=[]
    label=[]
    for _ in range(2):
        r=random.randint(0,len(fn)-1)
        inf=fn[r]
        (rate,sig)=wav.read(inf)
        mfcct=mfcc(sig,rate)
        data=data+list(mfcct)
        cat=int((inf.split("/")[-2]).split("_")[0])
        le=[]
        for i in range(round(6*len(mfcct)/l6)):
            le=le+[cat]
        label=label+le
    return np.array(data),label
```

```python
def make1(r):
    inf=fn[r]
    (rate,sig)=wav.read(inf)
    mfcct=mfcc(sig,rate)
    data=list(mfcct)
    cat=int((inf.split("/")[-2]).split("_")[0])
    le=[]
    for i in range(round(6*len(mfcct)/l6)):
        le=le+[cat]
    label=le
    return np.array(data),label
```

```python
def makedata():
    data=[]
    for _ in range(3500):
        data=data+[make3()]
    for _ in range(1500):
        data=data+[make2()]
    for _ in range(len(fn)):
        data=data+[make1(_)]
    random.shuffle(data)
    return data
```

```python
def  pad():
    D=[]
    m=max(len(x[0]) for x in dd)
    n=round(6*m/l6)
    for x in dd:
        x1=np.array(list(x[0])+(list(np.zeros([m-len(x[0]),13]))))
        labellength=n-len(x[1])
        if labellength>0:
            y1=x[1]+list(np.zeros([labellength]))
        D=D+[[x1,y1]]
    return D
```
